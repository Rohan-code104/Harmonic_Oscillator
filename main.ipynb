In this notebook, we apply Tao Xin et al.'s 3-step algorithm framework to the differential equation governing the classic harmonic oscillator:

\begin{equation*}
y'' + y = 0 \quad \Rightarrow \quad M = \begin{bmatrix} 0 & 1 \\ -1 & 0 \end{bmatrix}, \quad \vec{b} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \quad \vec{x}(0) = \begin{bmatrix} 1 \\ 1 \end{bmatrix}
\end{equation*}


Evaluating Kinetic and Potential Energy \\

Given m = 1 ; w = 1
\begin{equation*}
K(t) = \frac{1}{2}(y'(t))^2
\end{equation*}

\begin{equation*}
V(t) = \frac{1}{2}(y(t))^2
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt

# Mock Classiq functions (temporary)
class MockCircuit:
    def __init__(self, model):
        self.model = model

class MockJob:
    def result(self):
        return type("Result", (object,), {"value": type("Value", (object,), {"counts": {"000": 1000, "001": 1048}})})

def execute(circuit, shots):
    print(f"Executing circuit: {circuit.model}")
    return MockJob()

# Define time points for the evolution
time = np.linspace(0, 1, 11)  # 11 time points from 0 to 1

# Analytical solutions for Kinetic and Potential Energy
KE_analytical_time = [0.5 * (1 - np.sin(2 * t)) for t in time]
PE_analytical_time = [0.5 * (1 + np.sin(2 * t)) for t in time]

# Placeholder lists for quantum results
KE_quantum = []
PE_quantum = []

# Mock quantum model preparation and execution function
def prepare_and_execute_model(t):
    model = f"Quantum Model at t={t}"  # Mock model
    circuit = MockCircuit(model)  # Mock circuit
    job = execute(circuit, shots=2048)
    results = job.result().value.counts

    KE_sum = sum(v for k, v in results.items() if k.endswith('1')) / 2048
    PE_sum = sum(v for k, v in results.items() if k.endswith('0')) / 2048

    return KE_sum, PE_sum

# Run simulation for all time points
for t in time:
    KE, PE = prepare_and_execute_model(t)
    KE_quantum.append(KE)
    PE_quantum.append(PE)

# Plot comparison of K.E. (Quantum vs Analytical)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Kinetic Energy Plot
ax1.plot(time, KE_quantum, 'o-', color='blue', label="Quantum Solution (K.E.)")
ax1.plot(time, KE_analytical_time, 'x--', color='black', label="Analytical Solution (K.E.)")
ax1.set_title("Kinetic Energy Comparison: Quantum vs Analytical", fontsize=14)
ax1.set_xlabel("Time (T)", fontsize=12)
ax1.set_ylabel("Kinetic Energy", fontsize=12)
ax1.legend(loc="best")
ax1.grid(True)

# Potential Energy Plot
ax2.plot(time, PE_quantum, 'o-', color='green', label="Quantum Solution (P.E.)")
ax2.plot(time, PE_analytical_time, 'x--', color='black', label="Analytical Solution (P.E.)")
ax2.set_title("Potential Energy Comparison: Quantum vs Analytical", fontsize=14)
ax2.set_xlabel("Time (T)", fontsize=12)
ax2.set_ylabel("Potential Energy", fontsize=12)
ax2.legend(loc="best")
ax2.grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
